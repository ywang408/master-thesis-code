```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
```

```python
files = glob.glob("*.csv")
```

```python
data = pd.read_csv(files[0])
```

```python
v = np.arange(0.1,0.4,0.025)
data.iloc[:,1:]
v = pd.DataFrame({'vol':v})
data.iloc[:,2:].sub(data.mc,axis=0)
```

```python
def plot_price(path):
    df = pd.read_csv(path)
    df = df.iloc[:,1:]
    v = np.arange(0.1,0.4,0.025)
    v = pd.DataFrame({'vol':v})
    df = v.join(df)
    print(df.to_latex(index=False, caption=path))
    plt.figure(1)
    plt.ylabel('option price')
    plt.xlabel('vol')
    plt.plot(df.vol, df.mc, label='monte carlo result')
    plt.scatter(df.vol, df.w4, label='expansion result with N=4',color='red')
    plt.legend()
    df2 = df.iloc[:,2:].sub(df.mc, axis=0)
    df2 = df2.div(df.mc,axis=0)*100
    plt.figure(2)
    plt.title('error plot')
    plt.ylabel('error percentage')
    plt.xlabel('vol')
    plt.plot(df.vol, df2.w1, label='N=1')
    plt.plot(df.vol, df2.w2, label='N=2')
    plt.plot(df.vol, df2.w3, label='N=3')
    plt.plot(df.vol, df2.w4, label='N=4')
    plt.legend()
    plt.show()
```

```python
print(files[0])
plot_price(files[0])
```
